#### #8. Create a new column `pos_as_num` by using the mapping dictionary `{ C:5, G:1, F:3 }`. Then create a scatter matrix plot with the `pos_as_num`, `pts`, `age`, and `fg` columns.

(Hint: You'll have to do the same trick with `map` as we did in class with converting the type of avocado to a number.)

```python
mapping_dict = {}
initial_class_label = 0
for type in list(nba['pos'].unique()):
    mapping_dict[type] = initial_class_label
    initial_class_label += 1
mapping_dict
nba['pos_as_num'] = nba['pos'].map(mapping_dict)
nba[['pos', 'pos_as_num']].head(3)

nba_scatter = nba[['pos_as_num', 'pts', 'age', 'fg']]

pd.plotting.scatter_matrix(
    nba_scatter,
    figsize=(12,12)
);
```
